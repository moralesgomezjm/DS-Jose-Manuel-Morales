```python
import numpy as np

alumnos_aidone = np.array([
    [5, 4],
    [2, -3],
    [3, 5],
    [-5, 4],
    [0, -1],
    [3, 2],
    [-1, 1],
    [5, -2],
    [-5, 3],
    [-5, 1]
])
```

```python
teresa = [-2, -5]
udo = [0, 4]

nuevos = np.array([teresa, udo])
alumnos_concatenados_1 = np.concatenate((alumnos_aidone, nuevos), axis=0)

alumnos_concatenados_2 = np.vstack((alumnos_aidone, nuevos))
```

```python
alumnos_ref = ["Rodrigo", "Gabriela", "Mateo"]
lista_alumnos = [
    "Rodrigo", "Lucia", "Alejandro", "Valeria", "Javier", "Camila",
    "Diego", "Gabriela", "Mateo", "Sofía", "Teresa", "Udo"
]
curso_asignado = ["CAI", "ER", "NMPC"]
```

```python
num_alumnos = len(lista_alumnos)
matriz_reparto = np.zeros((num_alumnos, 3))
```

```python
for i_vector in range(num_alumnos):
    alumno = lista_alumnos[i_vector]
    vector_alumno = alumnos_concatenados_1[i_vector]

    print(f"\nPara {alumno} con vector {vector_alumno}:")
    vector_distancias = np.zeros(3)

    for indice_curso, alumno_ref in enumerate(alumnos_ref):
        indice_alumno_ref = lista_alumnos.index(alumno_ref)
        vec_ref = alumnos_concatenados_1[indice_alumno_ref]

        
        distancia = np.linalg.norm(vector_alumno - vec_ref)
        vector_distancias[indice_curso] = distancia

 
    matriz_reparto[i_vector, :] = vector_distancias


    indice_curso_elegido = np.argmin(vector_distancias)
    print(f"Clase asignada: {curso_asignado[indice_curso_elegido]}")
```

```python
import random

def colocar_barco_aleatorio(tablero, eslora=4):
    filas, cols = tablero.shape
    orientaciones = ['N', 'S', 'E', 'O']
    colocado = False

    while not colocado:
        fila_ini = random.randint(0, filas - 1)
        col_ini = random.randint(0, cols - 1)
        orient = random.choice(orientaciones)

        
        if orient == 'E':
            posiciones = [(fila_ini, col_ini + i) for i in range(eslora)]
        elif orient == 'O':
            posiciones = [(fila_ini, col_ini - i) for i in range(eslora)]
        elif orient == 'S':
            posiciones = [(fila_ini + i, col_ini) for i in range(eslora)]
        else:  # N
            posiciones = [(fila_ini - i, col_ini) for i in range(eslora)]

        
        dentro = all(0 <= f < filas and 0 <= c < cols for f, c in posiciones)
        if not dentro:
            continue

        
        libre = all(tablero[f, c] == " " for f, c in posiciones)
        if libre:
            for f, c in posiciones:
                tablero[f, c] = "O"
            colocado = True

    return tablero

tablero = colocar_barco_aleatorio(tablero)
print(tablero)
```

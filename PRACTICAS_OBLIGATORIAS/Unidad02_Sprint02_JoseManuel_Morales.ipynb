```python
class User:
    def __init__(self, nombre, saldo_inicial, checking_account):
        if saldo_inicial < 0:
            raise ValueError("El saldo inicial no puede ser negativo.")

        saldo_inicial = int(saldo_inicial)

        self.nombre = nombre
        self.saldo = saldo_inicial
        self.checking_account = bool(checking_account)

    def withdraw(self, cantidad):
        cantidad = int(cantidad)

        if cantidad < 0:
            raise ValueError("La cantidad no puede ser negativa.")

        if self.saldo < cantidad:
            raise ValueError("Fondos insuficientes.")

        self.saldo -= cantidad
        return f"{self.nombre} tiene ahora {self.saldo}."

    def deposit(self, cantidad):
        cantidad = int(cantidad)

        if cantidad < 0:
            raise ValueError("La cantidad no puede ser negativa.")

        self.saldo += cantidad
        return f"{self.nombre} tiene ahora {self.saldo}."

    def transfer(self, otro_usuario, cantidad):
        cantidad = int(cantidad)

        if cantidad < 0:
            raise ValueError("La cantidad no puede ser negativa.")

        if self.saldo < cantidad:
            raise ValueError("Fondos insuficientes para transferir.")

        if not otro_usuario.checking_account:
            raise ValueError("El usuario receptor no puede recibir transferencias.")

        
        self.saldo -= cantidad
        otro_usuario.saldo += cantidad

        return (f"{self.nombre} tiene ahora {self.saldo}. "
                f"Transferencia completada a {otro_usuario.nombre}, "
                f"quien tiene ahora {otro_usuario.saldo}.")

```

# EJERCICIO 2 #

```python
import numpy as np
```

```python
generos = ["accion", "aventura", "animacion","comedia","romance"]
peliculas = {"toy story":[0,1,1,1,0], "jumanji":[0,1,0,0,0], "mortal_kombat":[1,1,0,0,0], "pocahontas":[0,0,1,0,1]}

matriz = np.array(list(peliculas.values()))

print(" Matriz de películas:\n", matriz)
print("\nDimensión:", matriz.ndim)
print("Shape:", matriz.shape)
print("Tamaño total:", matriz.size)
print("Tipo:", type(matriz))
```

```python
def similitud_coseno(v1, v2):
    return np.dot(v1, v2) / (np.linalg.norm(v1) * np.linalg.norm(v2))
```

```python
nombres = list(peliculas.keys())

print("\nSimilitud entre películas:\n")
for i in range(len(nombres)):
    for j in range(i+1, len(nombres)):
        sim = similitud_coseno(matriz[i], matriz[j])
        print(f"{nombres[i]} , {nombres[j]}: {sim:.2f}")
```

```python
wall_e = np.array([1, 1, 0, 1, 0])

print("\n Similitud de Wall-E con las demás:\n")
for i, nombre in enumerate(nombres):
    sim = similitud_coseno(wall_e, matriz[i])
    print(f"Wall-E ↔ {nombre}: {sim:.2f}")
```

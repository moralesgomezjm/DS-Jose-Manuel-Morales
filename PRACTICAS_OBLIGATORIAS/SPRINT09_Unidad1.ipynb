```python

import pandas as pd
import numpy as np
from scipy import stats
```

```python
#dataset
titanic = pd.read_csv("titanic.csv")

# Vistazo
titanic.head()
```

```python
#Prob. Supervicencia de un niño
children = titanic[titanic["age"] < 18]

prob_survive_child = children["survived"].mean()
prob_survive_child
```

```python
#probabilidad de que una mujer viaje sola

women = titanic[titanic["sex"] == "female"]

prob_woman_alone = (women["alone"] == True).mean()
prob_woman_alone
```

```python
#probabilidad de sobrevivir un hombre de 30-40 años, en southampton y sin pagar billete

subset = titanic[
    (titanic["sex"] == "male") &
    (titanic["age"].between(30, 40)) &
    (titanic["embarked"] == "S") &
    (titanic["fare"] == 0)
]

prob_survive = subset["survived"].mean()
prob_survive
```

```python
#y acompañado de 2 familiares

subset_family = subset[(subset["sibsp"] + subset["parch"]) == 2]

prob_survive_family = subset_family["survived"].mean()
prob_survive_family
```

```python
#Intervalo 95% para supervivencia viajando solo

alone = titanic[titanic["alone"] == True]

p = alone["survived"].mean()
n = len(alone)

z = 1.96
se = np.sqrt((p * (1 - p)) / n)

ci_lower = p - z * se
ci_upper = p + z * se

(ci_lower, ci_upper)
```

```python
#solo en 2 o 3 clase

subset_extra = titanic[
    (titanic["class"].isin(["Second", "Third"])) &
    (titanic["alone"] == True)
]

p_extra = subset_extra["survived"].mean()
n_extra = len(subset_extra)

se_extra = np.sqrt((p_extra * (1 - p_extra)) / n_extra)

ci_lower_extra = p_extra - z * se_extra
ci_upper_extra = p_extra + z * se_extra

(ci_lower_extra, ci_upper_extra)
```

```python
import pandas as pd
from scipy import stats

# Cargar dataset
life = pd.read_csv("lifesat_full.csv")

# Comprobar que cargó bien
life.head()
```

```python
#LIFE SATISFACTION

#99% que la media es <7

mean = life["Life satisfaction"].mean()

t_stat, p_value = stats.ttest_1samp(life["Life satisfaction"], 7)
p_value_one_tail = p_value / 2

mean, p_value_one_tail
```

```python
#Nivel medio con vida >65 y empleo >60%

subset = life[
    (life["Life expectancy"] > 65) &
    (life["Employment rate"] > 60)
]

mean_subset = subset["Life satisfaction"].mean()
mean_subset
```
